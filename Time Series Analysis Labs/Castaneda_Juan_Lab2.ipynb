## General tips for computing and plotting discrete fourier transforms (DFT)

- Let g(t) be some time signal that’s sampled at $dt$ to get a discrete array/list `g = [ … ]`
- You don’t need to code your own DFT, use numpy:  `A = numpy.fft.fft(g) * dt`
- You should also use `numpy.fft.fftshift( A )` to shift the fft output such that the 0-frequency component is centered (see why here https://docs.scipy.org/doc/numpy/reference/routines.fft.html#background-information) which you probably want when plotting
- You can use `f_axis = numpy.fft.fftshift( numpy.fft.fftfreq(len(g), dt) )` to create the frequency axis for plotting the shifted spectrum
- Keep in mind the fft output is in general complex, so to compare two fourier transforms (e.g. DFT vs CFT) you should compare either the real and imaginary parts ($z = x + i y$), or, the phase and amplitude ($z = r e^{i\theta}$ ). Show all 4 aspects if you like.
- Note numpy fft assumes the time signal starts at $t = 0$, if yours doesn’t you should center it at zero. If you don’t then the complex components ($x$ and $y$, or phase) will be off, but the amplitude should not change (why? analytically, recall that axis shifts in either domain are equivalent to complex exponential scaling, which has amplitude 1, in the other domain).
- If your time signal $g$ is centered at zero, a trick to ‘rotate it’ to start at zero (and then take the fft and fftshift that) is to do: `fftshift( fft( ifftshift(g) ) )` (you may see mention of this online)

Convolving a general function, f(t), with a gaussian produces a "blurring" or "smoothing" effect, as each point in the original f(t) will essentially become a weighted average of the points around it under a convolution with a gaussian, g(t) (where the weights are given by g(t)).

If g(t) is narrow, only the closest points will be given any significant weight, whereas if g(t) is very wide, points that are far away may also contribute meaningfully to the weighted average. This means that the wider g(t) is (ie. the higher t_H is), the more "blurred" the result of the convolution will be.

This means that applying a gaussian function with t_H = 20 will have more of a smoothing effect than applying one with t_H = 20.

# 2 Fourier transform of Window Functions (5 pts)

A continuous time signal $f(t)$ can be truncated into a signal of finite length
$T$ by window functions $b(t)$:

$$ g(t) = f(t) b(t) $$

Typical window functions include:

 * Boxcar function
    $$b(t) =
    \begin{cases}
    1 & 0 \leq t \leq T
    \\
    0 & \text{else}
    \end{cases}
    $$
 * Hann window
    $$b(t) =
    \begin{cases}
    \frac{1}{2} \left( 1 - \cos \frac{2 \pi t}{T} \right) & 0 \leq t \leq T
    \\
    0 & \text{else}
    \end{cases}
    $$

# 2.1

```python
def Boxcar(t, T):
    return np.where(np.logical_and(t>=0, t<=T), 1, 0)


def Hann(t, T):
    return np.where(np.logical_and(t>=0, t<=T), 0.5*(1 - np.cos((2*np.pi*t)/T)), 0)


T = 10  # seconds
dt = 0.01  # seconds


t = np.arange(-2*T, 2*T, dt)

box = Boxcar(t, T)
hann = Hann(t, T)


plt.plot(t, box, label="Boxcar")
plt.plot(t, hann, label="Hann")
plt.title("Comparison of window functions")
plt.legend(loc="upper left")
plt.xlabel('t')
plt.ylabel('w(t)')
plt.show()
```

By convolution theorem:  ${\mathcal {F}}\{f*g\}={\mathcal {F}}\{f\}\cdot {\mathcal {F}}\{g\}$

This means that if we truncate some function f(t) with a window function in the original spectrum, and then use the FT to go to the frequency spectrum, the result will be the FT of f(t) multiplied by the FT of the window function. 

Thus, truncating with the boxcar will result in more "noise" in the frequency spectrum than truncating with the Hann function. This can be seen in the plots below.

The upside to the boxcar function is that it does not modify the original spectrum of the function like the Hann function does.

## Background

Liquids have no fixed internal structure. Yet they do have some short range order in the sense
that they have preferred intermolecular spacings, which are determined by the locations of minima
in the intermolecular potentials. The microscopic structure of liquids is often characterized by a
quantity known as the Radial Distribution Function $g(r)$, which is essentially the probability (Relative
to the average probability, which means that $g(r)$ tends to 1 at large $r$, where the neighbour is too far
away to feel any interaction.) that a molecule has a neighbouring molecule at distance $r$. Typically $g(r)$
shows a value that approaches zero at small $r$ since molecules cannot occupy the same space; it also shows a
peak at the preferred distance of nearest neighbours, and secondary peaks at preferred distances of more distant
neighbours. If a suitable collimated beam of particles (e.g. X-rays or neutrons) is sent through a
sample of the liquid, some of the particles are scattered. The number of particles scattered through
a given angle is related to the Fourier Transform of $g(r)$ evaluated at the wavenumber $k$
corresponding to the transverse momentum transfer associated with that scattering angle. Kittel derives
this relationship in Chapter 17 of Introduction to Solid State Physics.

If this all sounds complicated, all you need to know here is that something called the Structure
Factor $S(k)$ is effectively measured by looking at the scattered intensity as a function of
scattering transverse wavenumber $k$ (proportional to scattering angle), and that the Radial Distribution
Function is related to it by

$$ g(r) = 1 + \frac{1}{2 \pi^2 \rho r} \int_0^\infty k \left( S(k) - 1 \right) \sin(kr) dk $$

where $\rho$ is liquid number density (number of atoms per unit volume, computable from the three
constants mentioned in the introduction), $k$ is wavenumber, and $r$ is radius.

```python
def RDFcalc(S, dk, rho, k):
    # get the fourier transform of p using S and k (ie. P(k))
    P = (np.pi*k/1j)*(S - 1)  # P(k)   

    # shift so that values for negative k are at the end of the array (ie. the first element corresponds to k=0, to use with ifft)
    P = np.fft.ifftshift(P)

    
    # Calculate dr
    dr = 2*np.pi/(k[-1]-k[0])
    print('dk:', dk)
    print('dr:', dr)

    
    # do the ifft to get p(r)
    p = np.fft.fftshift(np.fft.ifft(P))/(dr*len(P))
 
   
    
    # change from k axis to r axis
    rn = np.fft.fftshift(np.fft.fftfreq(len(P), dk/(2*np.pi)))
    # note that this factor of 2pi is needed, similarly to question 1, since k is analogous to angular frequency here
    # so we need this factor of 2pi to correct the axis
    
    
    # calculate gn
    gn =  ((2*np.pi)**2) * np.real(p)/(np.pi*rho*rn)    +   1
    
    # Note: I noticed my answer was off with regards to scale, and multiplying the integral term in g(r) by (2pi)^2 corrected it
    #       exactly. I believe that numpy uses frequency, f, for the fourier transform, but k is essentially the angular
    #       frequency, w = 2pi*f. Therefore, k would be replaced by 2pi*k, and dk would be replaced with 2pi*dk, which accounts
    #       for this extra factor of (2pi)^2 in the equation above
    
    
    
    return gn, rn, p, dr


# convert rho to the proper units
rho = (massRho/molWeight)*Navogadro/(1e24)  # (g/cc)*(mol/g)*(atoms/mol)*(cc/angstrom) = atoms/cubic_angstrom  
print("rho:", rho)

gn, rn, p, dr = RDFcalc(S, dk, rho, k)


# Look at the graph of p(r) to see if ifft was done correctly
print("As a check to see if I did the ifft and set the r axis correctly, I look to see if p(r) is odd and real, which it is.")
plt.plot(rn, np.real(p))
plt.title("real part of p")
plt.show()
plt.plot(rn, np.imag(p))
plt.title("imaginary part of p")
plt.show()
print("Notice that the imaginary component is essentially zero when compared to the real component, as expected")


```

# Summary of fftshift and ifftshift

### To go from P to p to P (imag odd, real odd, imag odd):
1) ifftshift(P) to put the negative frequencies at the end

2) p = ifft(P) to get the ifft

3) p = fftshift(p) to center it for plotting

4) to go back to P we need to do p = ifftshift(p) again so it has negative values at the end

5) P = fft(p)

6) P = fftshift(P)


### summary: fftshift centers the array for plotting, ifftshift puts it back into the form where we can do calculations (fft and ifft) on it
